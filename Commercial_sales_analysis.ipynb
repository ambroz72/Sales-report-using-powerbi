```python
import matplotlib.pyplot as plt
```

```python
#Sales by order Date
```

```python
# Ensure the date column is in datetime format (replace 'Order Date' with the actual column name)
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group sales by date (replace 'Sales' with the actual column name)
sales_by_date = data.groupby('Order Date')['Sales'].sum().reset_index()

# Plot the line chart
plt.figure(figsize=(10, 5))
plt.plot(sales_by_date['Order Date'], sales_by_date['Sales'], marker='o', linestyle='-')

# Formatting the chart
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.title('Sales Trend Over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Profit by region
```

```python
# Group by 'Region' and sum the 'Profit'
profit_by_region = data.groupby('Region')['Profit'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(profit_by_region['Region'], profit_by_region['Profit'], color='skyblue', edgecolor='black')

# Formatting the chart
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit by Region')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability

# Show the plot
plt.show()
```

```python
# Group by 'Segment' and sum the 'Sales'
sales_by_segment = data.groupby('Segment')['Sales'].sum()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(sales_by_segment, labels=sales_by_segment.index, autopct='%1.1f%%',
        colors=['lightblue', 'lightcoral', 'lightgreen', 'gold'], startangle=140,
        wedgeprops={'edgecolor': 'black'})

# Formatting the chart
plt.title('Sales Distribution by Segment')

# Show the plot
plt.show()
```

```python
# Group by 'Region' and sum the 'Sales'
sales_by_region = data.groupby('Region')['Sales'].sum()

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%',
        colors=['blue', 'coral', 'green', 'gold', 'pink'],
        startangle=140, wedgeprops={'edgecolor': 'black'})

# Adding a white circle at the center to create the "Donut" effect
centre_circle = plt.Circle((0, 0), 0.6, fc='white')
plt.gca().add_artist(centre_circle)

# Formatting the chart
plt.title('Sales Distribution by Region')

# Show the plot
plt.show()
```

```python
# Group by 'Sub-Category' and sum the 'Sales'
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()

# Sort values in descending order (important for funnel shape)
sales_by_subcategory = sales_by_subcategory.sort_values(by='Sales', ascending=False)

# Create a "Funnel" effect using varying bar widths
plt.figure(figsize=(10, 6))

# Normalize bar widths for a funnel effect
max_width = 0.8  # Maximum bar width
min_width = 0.3  # Minimum bar width
bar_widths = np.linspace(max_width, min_width, len(sales_by_subcategory))

# Plotting the funnel chart
for i, (subcategory, sales) in enumerate(zip(sales_by_subcategory['Sub-Category'], sales_by_subcategory['Sales'])):
    plt.barh(subcategory, sales, color='blue', edgecolor='black', height=bar_widths[i])

# Formatting the chart
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.title('Sales horizontal bar chart by Sub-Category')
plt.gca().invert_yaxis()  # Invert y-axis to start from the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
